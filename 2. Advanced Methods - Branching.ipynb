## Branching Concepts

____

We started off by working or been in the master branch. Mastter is the main or the default branch in a repo. So when you start everything will be in this branch.

So we started off with this, where we are committing our changes to the mster branch:

![18-Branching1.png](attachment:18-Bra nching1.png)


But what it we, instead of committing to the master branch, we

1. Create another branch, e.g. lets call it a feature branch
2. And commit to that branch




But what happens if we create another branch?


So master and feaure branch will be the same at first, since feature is a product of master. But, since feature, is a tributary of master, when changes are committed to feature (and does not to master), those changes will only be available in the feature branch.

And lets say we switchback to the master branch, then the changes we made to the feature branch will not be available in master branch (correct) (...we just switched back to master remember)


Each commit only knows what happens in its own branch.


Why is this (branching) useful? Its very useful, as you may want to add new features to an already running or existing program, and by doing so, you dont want to introduce breaking code, to the existing program.  So you want to work in some kind of a sandbox area, so when we happy it works properly, we can merge it back to the main branch (and pray it will work...with the old code...this is another consideration one has to take into account:)


This is very helpful, when there are many people working in the same repositoy.




So we will be working on feature branch for weeks, as new feautures takes time. But what if there is an existing bug, in the master branch? We need to create a hotfix branch, to deal with this situation, luckily we only needed one commit, made sure it was working correcty on its own, then we merged it back to the master branch.












## Lets get practical at the command line

____

We will work on demo_repo

#### How to get the branch information

```
git branch
```

Shows we only have 1 branch

```
* master
```
And the * means that we are currently on that branch

then type ```q``` to exit

#### Create a new branch

We use the checkout command 

1. Create a new branch, with ```-b``` option
2. Switch between branches
3. Then specify the name, and attach some additional information related to the reason for the branch (```feature-POC976```)

```
git checkout -b feature-readme-update
```

So it creates the branch AND switches to it.

```
Switched to a new branch 'feature-readme-update'
```

And ```git branch``` will give us:
```
* feature-readme-update
  master

```

Meaning feature is the active branch/

And we can switch back (between branches) with

```
git checkout master
```
```
Your branch is up to date with 'origin/master'.
```
And we back to

```
  feature-readme-update
* master
```


Lets go back (tab auto complete)

```
git checkout feature-readme-update
```

##  Pull Request

____

What da heck is a PR? It is a request to have your code pulled (merged) into another branch.

Reason why it is pull? As The owner is the source or master, and he pulls into his branch. It cannot be push, as the coder (of new feauture) dont have freedom to do this (he is not owner). Its a MUST be a pulling action, as we bring into towards us (I am owner). Again cannot be push as coder (feature) do not have the permission to do so. Hope that is clear.

So once we made a PR:

1. Anyone ca review our code
2. Comment on it (to make changes or updated)

Also once you have made a PR:

1. You can make further changes
2. And do commits (additional)
3. And push it back into github

As only it is under the same branch.

Once the PR is merged, you will delete your your feature branch, and swicth back to master branch.

If you want to add more feauture, you will start the process (creating a new branch) over again.


Steps, summary

1. Create feature-branch
2. Check out feature-branch
3. Make changes in feature-branch
4. Add, commit and paush to github

## Merge Conflicts

____

Ok, so far its been plain sailing, but life is never so uneventful. E.g. I am writing code on my branch, and other people are working on their branch.

And remember master is getting updated from different places. So it is possible for

1. Multiple people to change the same files

And got does not know which code you want to keep or throw away.

This requires manual intervention.


#### Lets create another branch

```
git checkout -b quick-test

```

Modify index.html (<p>World</p>)... add and commit it

We use the ```-am``` for add and commit, with m for message, at the same time. (only works for modified files, not newly created files)

```
git commit -am "Updated Index.html"

```

Lets go to master

```
git checkout master
```
and we should not have the changes in index.html. But what If I updated the this file under the master branch???

But here we add text <p>There</p>


What doe we have now>

1. master has change in index.html (...)
2. quick-test has change in index.html (...) 

Separately they fine, but when merge happens is it still a pull to master, meaning master will be overwritten with feature (or quick-test, in our case)

Lets see:

Checkout quick-test

```
git checkout quick-test
```


Now we getting an error...

It does not allow me to change ny branch, as I have local changes, meaning changes in master branch, that is not in quick-test branch. It saying that the master copy will loose its changes if I change branch or overwritten, and that I first need to commit this change in master, and will then not be lost when I change branches.


It says commit or stash your changes? What is stashing? Means stash the change somewhere and retrieve them later, but it is not making a commit to git. Like a temporary holding place.


Lets commit

```
git commit -am "Added there"
```

Now we should be able to check out

```
git checkout quick-test

```


Lets merge to master. Remember we said that this (merging locally) is not the preferred pattern (```merge master```), rather to create a pull request.

It is not the preferred pattern for merging branches into master, but master get updated as you work (and others also work, that is updating or merging into master), and you dont have those changes in your branch. So you dont want to get too far behind, as you working. So you going to want to pull those changes down to your local master. So you want your branch (quick-test) to be kept up to date, what is going on in master. 

So this is the reason for

(must be in our quick-test branch)
```
git merge master
```




Now we get a conflict. 




How do we solve this?

1. VSCode gives an interface to fix it (easiest way to fix it)
2. GitHub gives an interface to fix it
3. But can also fix it in the terminal

We will do point 1 above

1. We can click on any of those options
2. But we can also make the change in the code itself, by deleting the lines that git added

How do we read this?

We on branch, quick-test, so this shows us what we have done in this branch 

```
<<<<<<< HEAD (Current Change)
<p>World</p>
````

The below shows what is in the master branch, but not in quick-test branch
```
<p>There</p>
>>>>>>> master (Incoming Change)

````

So we do the following, by deleting git added stuff:

```
<div>Hello</div>
<p>World</p>
<p>There</p>
```

Or click on ```Accept both changes``` does the same thing.

And if we do git status, we can see we need to do another commit




We need to do another commit, as we have fixed the merge conflicts


```
git commit -am 'updated with master`
```

So we have updated out branch with masters changes,and continue coding on our feature
