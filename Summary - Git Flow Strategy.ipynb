# Git Flow Strategy

____

Git flow is an example of a branching strategy. 

Branching is used to define the workflow of software development. Can be
- simple 
- complex (branch heavy)



Git Flow is a a workflow strategy that revolves around long lived branches.

We can have a development branch, where we do concentrate on development, but we can also do releases of new features, and still be able to do current releases.

Two primary branches

- Develop branch
- Master branch

Master as all times will have production code in it, while development is copy of master, but also the latest development of the next release.

From the main and develop we have supporting branches.


From Develop we will have the following branches
- Feature
- Release

A release can also be called a sprint (just shoter time period. Also sounds like a featur)



Release branches from Develop, which is new code, that is under development that will be released in next version. so the flow is:
- Develop->branch to Release->Add Changes->Merge with Develop and Main->Delete Release

Feature branches from Develop, which is new features that is under development. The flow is as follows:
- Develop->branch to Features-Add new code->Merge with Develop


From Main we will have the following branches
- Hotfix

Hotix branches from Master, which involves fixing critical production code, under the master branch. It will not impact code under development, but will be merge with development and master when done. The flow is
- Master->branch to Hotfix->Add changes->Merge with Develop and Master-Delete Hotfix






# Making a new  Release

___

In order to add new features, we make a feature branch from the develop branch. And can have multiple feaures from the develop branch. You will then decide which feature you need to merge with develop branchg. Once yo have merged the feature(s) to the develop branch, you create a release branch.

Now you can continue with work on the development branch, without disruption the release of our work, that will be going to be implemented in production.

Remember we have Release and Develop branch now. There may be, in fact there will be bugs in the releases branch, which needs to be fixed, provided it is small fixes. No major work/changes must be carried out in this branch (Save these for a future release). So this is fine, and we accept that development and release will be out of sync. Thats fine.

Release is ready, we merge the release with the master, and tag it with a release number (v1.2)






Release branch, with bug fixes, is then merged into Develop. So master and Develop should be in sync, provided there where no features merged after the released was brached off. 



# Dealing with Hotfixes
___

Lets say we have a critical bug in production, we will make a hotfix branch from master, fix the bug in hotfix . 

Remember planned work in production still continues, so there is no impact under development.

Now bug is fixed, merge the hotfix back into master, Tag the hotfix, with version number (v1.3)

But also merge the hotfix back into develop or release branch, if there is one.


# Typical Flow

### First time getting code

We have the following components or elements in git

- Local workspace or folder
- Index, staging area, where we define or tag files that we want to push to remote
- HEAD, local repository (existng, plus tagged files), these are the files that is ready (tagged files above) to go to local reposity
- remote repository



We start off getting a copy of the project, by cloning the repository from gitlab into your workspace.

```
git clone 
```

We add new files to our workspace. Next we decide which files that we added to the workspace/folder, needs to go to the reposity. And for that we use the ```add``` command

```
git add textfile.txt
```

We can also specify all the files with

```
git add .
```

Those files will now be added to our staging area, which is a logical structure in the .git (hidden ) folder structure on your machine.

Those files are marked to be committed, not yet committed.

Next step is to commit it, to out local repository or HEAD (git terminology). This is the package that will be uploaded to git

```
git commit -m'Feaure 001.added'

```

Can also to staging and commit with one command. But this will NOT include new files, to be commited to HEAD. Only files that was already in repo (only changed files!)

```
git commit -a [does the add and commit at the same time]
```

Next we need to push it to remote reposity, we will use the push command


```
git push -u master 
```

Lets reverse the process


### Aready have the project on machine, get latest code.


We fetch the new code from repositoty to you local repo. Remember its only in your local repo, NOT in working folder workspace. So you will not see the files. What use is this for you?


```
git fetch 
``` 

To get the files to local workspace, we issue a merge command

```
git merge
```
But we can combine the above two commands with single command, which is pull

```
git pull
```




The reason why we have fetch and merge process, two distinct processes, is to allow us to check the differences between our workspace and remote repository content (which is now available in our local repository)


The command we used is ```diff head```. And decide whether I want to merge or not

```
git diff head
```

This is probably how all will do it, since the reposity may have code from other changes (especially files that you have changed on your machine),that you do not have in your workspace, and you have also worked on stuff, which is not on remote repository.

We can also compare what is in workspace, with files that are in staging area, reading to be committed. We dont use the head command. This is ideal way to see that the chanhes you have made, if they have been staged, i.e. ready to be committed.

```
git diff
```


## Practical Example

https://www.youtube.com/watch?v=9pa_PV2LUlw [very good]
