
The main idea with continuous integration is to deal with new code that is created
- through new branch
- where features have been added.

And to automate REVIEWING of this code. It is an automated reviewing or verification process against the master. If CI is setup, then the pushing of code to github, is done in backend, even before the pull request is created by the developer.

When the developer adds new code, it will be done as follows:
1. Adds a new branch locally on the machine [by checkout -b option]
```
git checkout -b

```
2. Codes the new feautures
3. Add and commits it
4. Push the new branch to github
5. Or create the branch when pushing to github (shorter option, instead of creating branch locally. Except with point 5, the branch is only on github, NOT locally)

```
git push -u orgiin main: feature-003
```

With Circle  (Summary)
1. Sign up
2. Through github
3. Select project you want to integrate
4. Create config.yml file (or select their one. Good starter
5. (Select pull request of this project, to create ```.circleci/config.yml```)
6. Now is project is setup to listen for any new push actions from developer.

When developer push a new branch, circle ci will automatically start running test, checking it agaist main etc. This is even before creating a pull request.

When you do create a pull request, then CI is already done, as it occurs on the push (check the CircleCI dashboard).
