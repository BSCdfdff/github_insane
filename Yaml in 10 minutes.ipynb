object:
    key:{space}value
```
E.g
```
holidayReconConfig:
    destDir: dst/
```

RULES 2 : Object and list(item), with attribute of the list on same line. No space after the hyphen and listitem. Very important. And hyphen, start at the same column as the previous attrubute.There CAN be space, ONLY if the list item is the first element
```
object:
    -listitem:{space}value
```

E.g.

```
holidayReconConfig:
    -system: generic    
```

RULES 3 : Object and list(item), with NO attribute on the same line, and the next attribute start sat the beginning column of the listitem text, NOT at the begining of the listitem hyphen. But if it does start at the hyphen, its a sibling, not attribute

```
object:
    -listitem:
    {space}attribute #starts at the same column as the listitem
```
E.g

```
holidayReconConfig:
    -system: 
     generic 
```

RULES 4 :  Object and list(item), with attribute of the list on same line. No space after the hyphen and listitem. With atribute at next line (which key and value pair). There must be no space between the key and value pair.

```
object:
    -listitem:{space}value
    {space}attribute(key):value #no space between key and value
```
E.g

```
holidayReconConfig:
    -system: 
     generic 
```





### Again

The flow has to be Object always first...

And before starting a list, it must again jhave parent, wihout attribute

Also do not think JSON. YAML is a slighly different way of thinking.

WORKS
```
