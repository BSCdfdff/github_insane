## Undoing

____



Lets say we made a mistake, and we committed it already. We can undo our commits (or stages)


Lets make change in readme.md (```2. Have Fun```), under quick-test branch.


Running git status, we can see we need to commit



Lets first add

```
git add README.md
```

So we have just staged it.



But we made a mistake, and want to remove it from ```add stage```?

#### Undo using reset for stage
(with no arguments or name of file)

```
git reset README.md
```

Now it is unstaged.



How do I undo a commit (previous was an ```add```)?
(lets first make the mistake)

```
git add .
git commit -m "Added have fun"
```
#### Undo using reset for commit
HEAD is pointer to last commit

And ~1, one back from last commit (meaning readme.md we udpdate

```
git reset HEAD~1
```



Now it 

1. Unstaged
2. Uncommitted 

The changes

HEAD is the only pointer to the commit, so if we need to get specific or have multiple ones, there is not easy what. But we can issue


```
git log
```

Which presents the commits together with ist hashes.



And we can then use these hashes, when we want to undo
(remove there)

```
git reset 35b1b20e9b096ea63242e984ca2eb251db2c6252
```
So it will unstage the changes made to files...but changes are still there?? ("there" is still there), but just not saved in git. (or commited to git)

Lets say now, I want to get rid of changes after a certain point? Again we use the git reset command

Lets get a commit hash

```
git log
```

And we issue

hard meaning, we not only want the changes to be unstanged, BUT to be completely removed.



```
git reset --hard 35b1b20e9b096ea63242e984ca2eb251db2c6252
```


Now head wil point to different commit (as we have essentially undo several commits)

NOW the text in file is gone/updated.

(and we have gone back in time)
