## Forking
___



This makes a complete copy of the repository. But why would you want to fork? Firstly you wont want to fork your own repo, as you have ful access already.


But lets say we want to work on someone else's repo. We dont have access. And If I need to make a PR against this repor and request for my changes to be added,  or if I just want to be able to branch of the code in this repo, and do whatever I want with it, I need to fork it.

Lets hit fork button and fork (react). Now we have the following:



Now this repository is under my person github. And I have complete control to make any updates or changes.

Here we can say update the readme file, and commit the changes.


But what If I need to now get my changes into react repository that is owned by facebook?


I will then create a PR.


1. Click Pull Request
2. New Pull Request




So the branch I was making modifications on, will be merged with branch of the original repository.

Then click on pull request, which we will not do.


But what If I want to merge within my fork, we can also do this:



 
But can also merge into any other branch

So after the merge, we will clean up the branch, that has just been merged. That is we will delete it from Gihub.

So me having my own fork, allows me full access and control of CRUD operations.

Procedure

1. Make Fork
2. Pull the information to my local machine
3. Make major updates, test it
4. Push it to my fork
5. Create a PR, against the original repository
