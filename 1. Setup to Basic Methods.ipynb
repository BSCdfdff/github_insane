## What is Git

____



Free open source version control library, most widely used today

What is version control system? Manages the changes to documents and programs. It is a way we track our code changes.

And allows use to go back to previous versions of code.

We will be using the CLI extensively, to learn the git commands and workflows. We use the term repository to specify the folder where your project is kept on your computer. GitHub is website to host your repositories online, and makes it easy to work on groups. This is also called a remote repository.

#### push

#### pull

Lets say you made changes on the remote repository, and those changes are not on your local copy, then we issue this command. This is the opposite from push.
